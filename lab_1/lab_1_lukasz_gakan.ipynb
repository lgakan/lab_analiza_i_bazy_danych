```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def fun(x):
    return x**2 + 5


def print_chart(x1, color='b'):
    plt.plot(x1, fun(x1), color=color, label='f(x)')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("f(x) = x^2 + 5")
    plt.legend()
    plt.show()
```

```python
# zad 3
# Polegalo na zdefiniowaniu funkcji f(x)=x^2+5 oraz sporzadzeniu jej wykresu dla podanych przedzialow
x_1 = np.linspace(-1, 1, 100)
x_2 = np.linspace(-6, 6, 100)
x_3 = np.linspace(0, 5, 100)
print_chart(x_1, 'r')
print_chart(x_2, 'g')
print_chart(x_3, 'b')
```
