```python
# zad 7 - Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.
df = pd.read_sql("SELECT cu.first_name, cu.last_name, co.country " +
                 "FROM customer AS cu " + 
                 "INNER JOIN address AS ad " + 
                 "ON cu.address_id = ad.address_id " +
                 "INNER JOIN city AS ci " +
                 "ON ci.city_id = ad.city_id " +
                 "INNER JOIN country AS co " + 
                 "ON co.country_id = ci.country_id " + 
                 "WHERE co.country = 'Nigeria' " +
                 "OR co.country = 'Poland' " +
                 "OR co.country = 'Bangladesh'",con=connection)
df
```

```python
# zad 10 - Jakie kategorie filmów zostały wypożyczone przez klientów?
df = pd.read_sql("SELECT DISTINCT  ca.name " +
                 "FROM category AS ca " +
                 "INNER JOIN film_category AS fi_ca " +
                 "ON ca.category_id = fi_ca.category_id " +
                 "INNER JOIN film AS fi " +
                 "ON fi_ca.film_id = fi.film_id " + 
                 "INNER JOIN inventory AS inv " +
                 "ON fi.film_id = inv.film_id " +
                 "INNER JOIN rental AS re " +
                 "ON inv.inventory_id = re.inventory_id", con=connection)
df
```

```python
# zad 11 - Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.
df = pd.read_sql("SELECT DISTINCT ca.name, co.country " +
                 "FROM category AS ca " +
                 "INNER JOIN film_category AS fi_ca " +
                 "ON ca.category_id = fi_ca.category_id " +
                 "INNER JOIN film AS fi " +
                 "ON fi_ca.film_id = fi.film_id " + 
                 "INNER JOIN inventory AS inv " +
                 "ON fi.film_id = inv.film_id " +
                 "INNER JOIN rental AS re " +
                 "ON re.inventory_id = inv.inventory_id " +
                 "INNER JOIN customer AS cu " + 
                 "ON cu.customer_id = re.customer_id " +
                 "INNER JOIN address AS ad " + 
                 "ON cu.address_id = ad.address_id " +
                 "INNER JOIN city AS ci " +
                 "ON ci.city_id = ad.city_id " +
                 "INNER JOIN country AS co " + 
                 "ON co.country_id = ci.country_id " + 
                 "WHERE co.country = 'United States'", con=connection)
df
```
